```python
import pandas as pd

# Load dataset
df = pd.read_csv("student_marks_dataset.csv")

# Calculate average marks per subject
avg_marks = df[['Math', 'Science', 'English']].mean()

# Highest and lowest subject
highest_subject = avg_marks.idxmax()
lowest_subject = avg_marks.idxmin()

print("Average Marks per Subject:")
print(avg_marks)

print("\nHighest Scoring Subject:", highest_subject)
print("Lowest Scoring Subject:", lowest_subject)
```

```python
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```
